```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from AccidentInput import file_inputl, combine_data
import warnings
warnings.filterwarnings("ignore")
```

```python
import os

# Define directories
traindir = "C:/Users/jgh6ds/Desktop/Independent/Sessions 2-5 Data and Visualization/TrainData"
sourcedir = "C:/Users/jgh6ds/Desktop/Independent/Sessions 6-8 PCA"

acts = file_inputl(traindir)
comvar = set(acts[0].columns).intersection(set(acts[7].columns))
totacts = combine_data(acts, list(comvar))

# Get dimensions of the combined dataframe
print(totacts.shape)
```

```python
# For ACCDMG
sns.boxplot(y='ACCDMG', data=totacts)
plt.show()

# Find the upper whisker
upper = totacts['ACCDMG'].quantile(0.75) + 1.5 * (totacts['ACCDMG'].quantile(0.75) - totacts['ACCDMG'].quantile(0.25))
xdmg = totacts[totacts['ACCDMG'] > upper]

# For Casualties (TOTINJ + TOTKLD)
xdmg['Casualties'] = xdmg['TOTINJ'] + xdmg['TOTKLD']

# Investigate extremes among the extremes
extremes = xdmg[xdmg['ACCDMG'] > 15e6]
print(extremes)

# Remove specific extreme
xdmg = xdmg.drop(extremes.index[0])
```

```python
# Remove duplicates
xdmgnd = xdmg.drop_duplicates(subset=['INCDTNO', 'YEAR', 'MONTH', 'DAY', 'TIMEHR', 'TIMEMIN'])

# Reset row indices
xdmgnd.reset_index(drop=True, inplace=True)
```

```python
# Principal Components of metrics for extreme accidents
features = ['CARSDMG', 'EQPDMG', 'TRKDMG', 'ACCDMG', 'TOTKLD', 'TOTINJ']
x = xdmgnd[features]

# Standardize the data
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)

# Perform PCA
pca = PCA()
pca.fit(x_scaled)

# Biplot
def biplot(pca, x_scaled, labels):
    plt.figure(figsize=(10, 7))
    plt.scatter(x_scaled[:, 0], x_scaled[:, 1], alpha=0.2)
    for i, (x, y) in enumerate(zip(pca.components_[0], pca.components_[1])):
        plt.arrow(0, 0, x, y, color='r', alpha=0.5)
        plt.text(x, y, labels[i], color='g')
    plt.xlabel('PC1')
    plt.ylabel('PC2')
    plt.show()

biplot(pca, x_scaled, features)
```

```python
# (example indices used here)
outliers = [5843, 5289, 5844, 5282]
xdmgnd_no_outliers = xdmgnd.drop(index=outliers).reset_index(drop=True)
x_no_outliers = xdmgnd_no_outliers[features]

# Standardize the data
x_no_outliers_scaled = scaler.fit_transform(x_no_outliers)

# Perform PCA again
pca_no_outliers = PCA()
pca_no_outliers.fit(x_no_outliers_scaled)

# Biplot without outliers
biplot(pca_no_outliers, x_no_outliers_scaled, features)
```

```python
# Scree plot
plt.figure(figsize=(8, 5))
plt.plot(range(1, len(pca_no_outliers.explained_variance_ratio_) + 1), pca_no_outliers.explained_variance_ratio_, 'o-', linewidth=2)
plt.title('Scree Plot')
plt.xlabel('Principal Component')
plt.ylabel('Variance Explained')
plt.show()

# Cumulative variance
plt.figure(figsize=(8, 5))
plt.plot(range(1, len(pca_no_outliers.explained_variance_ratio_) + 1), 
         pca_no_outliers.explained_variance_ratio_.cumsum(), 'o-', linewidth=2)
plt.title('Cumulative Variance Explained')
plt.xlabel('Principal Component')
plt.ylabel('Cumulative Variance Explained')
plt.show()
```

```python
# Loadings plot
loadings = pd.DataFrame(pca_no_outliers.components_.T, columns=['PC' + str(i) for i in range(1, len(features) + 1)], index=features)

plt.figure(figsize=(12, 6))
sns.heatmap(loadings, annot=True, cmap='coolwarm')
plt.title('PCA Loadings')
plt.show()
```

```python
# Perform PCA with correlation matrix
pca_corr = PCA()
x_corr_scaled = StandardScaler().fit_transform(xdmgnd[features])
pca_corr.fit(x_corr_scaled)

# Biplot with correlation matrix
biplot(pca_corr, x_corr_scaled, features)
```

```python
# Biplot comparison side by side
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Biplot with covariance matrix
axes[0].scatter(x_no_outliers_scaled[:, 0], x_no_outliers_scaled[:, 1], alpha=0.2)
for i, (x, y) in enumerate(zip(pca_no_outliers.components_[0], pca_no_outliers.components_[1])):
    axes[0].arrow(0, 0, x, y, color='r', alpha=0.5)
    axes[0].text(x, y, features[i], color='g')
axes[0].set_xlabel('PC1')
axes[0].set_ylabel('PC2')
axes[0].set_title('Biplot with Covariance Matrix')

# Biplot with correlation matrix
axes[1].scatter(x_corr_scaled[:, 0], x_corr_scaled[:, 1], alpha=0.2)
for i, (x, y) in enumerate(zip(pca_corr.components_[0], pca_corr.components_[1])):
    axes[1].arrow(0, 0, x, y, color='r', alpha=0.5)
    axes[1].text(x, y, features[i], color='g')
axes[1].set_xlabel('PC1')
axes[1].set_ylabel('PC2')
axes[1].set_title('Biplot with Correlation Matrix')

plt.tight_layout()
plt.show()
```

```python
# Scree plot comparison
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Covariance matrix
axes[0].plot(range(1, len(pca_no_outliers.explained_variance_ratio_) + 1), pca_no_outliers.explained_variance_ratio_, 'o-', linewidth=2)
axes[0].set_title('Scree Plot with Covariance Matrix')
axes[0].set_xlabel('Principal Component')
axes[0].set_ylabel('Variance Explained')

# Correlation matrix
axes[1].plot(range(1, len(pca_corr.explained_variance_ratio_) + 1), pca_corr.explained_variance_ratio_, 'o-', linewidth=2)
axes[1].set_title('Scree Plot with Correlation Matrix')
axes[1].set_xlabel('Principal Component')
axes[1].set_ylabel('Variance Explained')

plt.tight_layout()
plt.show()

# Cumulative variance comparison
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Covariance matrix
axes[0].plot(range(1, len(pca_no_outliers.explained_variance_ratio_) + 1), pca_no_outliers.explained_variance_ratio_.cumsum(), 'o-', linewidth=2)
axes[0].set_title('Cumulative Variance Explained with Covariance Matrix')
axes[0].set_xlabel('Principal Component')
axes[0].set_ylabel('Cumulative Variance Explained')


# Correlation matrix
axes[1].plot(range(1, len(pca_corr.explained_variance_ratio_) + 1), pca_corr.explained_variance_ratio_.cumsum(), 'o-', linewidth=2)
axes[1].set_title('Cumulative Variance Explained with Correlation Matrix')
axes[1].set_xlabel('Principal Component')
axes[1].set_ylabel('Cumulative Variance Explained')

plt.tight_layout()
plt.show()
```

```python
# Loadings plot for correlation matrix
loadings_corr = pd.DataFrame(pca_corr.components_.T, columns=['PC' + str(i) for i in range(1, len(features) + 1)], index=features)

fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Covariance matrix loadings plot
sns.heatmap(loadings, annot=True, cmap='coolwarm', ax=axes[0])
axes[0].set_title('PCA Loadings (Covariance Matrix)')

# Correlation matrix loadings plot
sns.heatmap(loadings_corr, annot=True, cmap='coolwarm', ax=axes[1])
axes[1].set_title('PCA Loadings (Correlation Matrix)')

plt.tight_layout()
plt.show()
```

```python
# Scatterplot matrix (SPM) using seaborn
predictor_features = ['ACCDMG', 'TRNSPD', 'CARS', 'TIMEHR', 'TEMP']
sns.pairplot(xdmgnd[predictor_features])
plt.show()

# PCA for potential predictors
x_predictors = xdmgnd[predictor_features]
x_predictors_scaled = StandardScaler().fit_transform(x_predictors)

pred_pca = PCA()
pred_pca.fit(x_predictors_scaled)

# Biplot for predictors
biplot(pred_pca, x_predictors_scaled, predictor_features)
```
