```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.linear_model import LinearRegression
import numpy as np

import warnings
warnings.filterwarnings("ignore")
```

```python
import os
from AccidentInput import file_inputl, combine_data

# Define directories
traindir = "C:/Users/jgh6ds/Desktop/Independent/Sessions 2-5 Data and Visualization/TrainData"
sourcedir = "C:/Users/jgh6ds/Desktop/Independent/Session 9-12 Multiple Linear Regression"

acts = file_inputl(traindir)
comvar = set(acts[0].columns).intersection(set(acts[7].columns))
totacts = combine_data(acts, list(comvar))

# Get dimensions of the combined dataframe
print(totacts.shape)
```

```python
# Identify extreme accidents
dmg_threshold = totacts['ACCDMG'].quantile(0.75)
xdmg = totacts[totacts['ACCDMG'] > dmg_threshold]

# Remove duplicates
xdmgnd = xdmg.drop_duplicates(subset=['INCDTNO', 'YEAR', 'MONTH', 'DAY', 'TIMEHR', 'TIMEMIN'])
```

```python
# Create Cause variable
xdmgnd['Cause'] = np.where(xdmgnd['CAUSE'].str.startswith('M'), 'M', 
                           np.where(xdmgnd['CAUSE'].str.startswith('T'), 'T',
                           np.where(xdmgnd['CAUSE'].str.startswith('S'), 'S',
                           np.where(xdmgnd['CAUSE'].str.startswith('H'), 'H', 'E'))))

xdmgnd['Cause'] = xdmgnd['Cause'].astype('category')

# Create Type variable
type_labels = ["Derailment", "HeadOn", "Rearend", "Side", "Raking", "BrokenTrain", 
               "Hwy-Rail", "GradeX", "Obstruction", "Explosive", "Fire", "Other", "SeeNarrative"]
xdmgnd['Type'] = pd.Categorical.from_codes(xdmgnd['TYPE'] - 1, categories=type_labels)
```

```python
# Scatter plot matrices
sns.pairplot(xdmgnd[['ACCDMG', 'TRNSPD', 'CARS', 'TIMEHR', 'TEMP']])
plt.show()
```

```python
# PCA
pca = PCA(n_components=2)
pca.fit(xdmgnd[['ACCDMG', 'TRNSPD', 'CARS', 'TIMEHR', 'TEMP']])
pca_results = pca.transform(xdmgnd[['ACCDMG', 'TRNSPD', 'CARS', 'TIMEHR', 'TEMP']])

plt.figure(figsize=(10, 7))
plt.scatter(pca_results[:, 0], pca_results[:, 1], c=xdmgnd['ACCDMG'], cmap='viridis')
plt.colorbar(label='ACCDMG')
plt.xlabel('PCA1')
plt.ylabel('PCA2')
plt.title('PCA Biplot')
plt.show()
```

```python
# Heatmap
cause_type_table = pd.crosstab(xdmgnd['Cause'], xdmgnd['Type'])
sns.heatmap(cause_type_table, annot=True, fmt="d", cmap="YlGnBu")
plt.title('No. of Accidents by Cause and Type of Accident')
plt.show()

# Interaction plots
sns.lmplot(x='TRNSPD', y='ACCDMG', col='Type', data=xdmgnd, logx=True, col_wrap=4, scatter_kws={"s": 10}, ci=None)
sns.lmplot(x='TRNSPD', y='ACCDMG', col='Cause', data=xdmgnd, logx=True, col_wrap=4, scatter_kws={"s": 10}, ci=None)
```

```python
# Single predictor
lm1 = LinearRegression()
lm1.fit(xdmgnd[['TEMP']], xdmgnd['ACCDMG'])
print(f'Coefficients: {lm1.coef_}')
print(f'Intercept: {lm1.intercept_}')
print(f'Residual sum of squares: {np.sum((lm1.predict(xdmgnd[["TEMP"]]) - xdmgnd["ACCDMG"]) ** 2)}')

# Two predictors
lm2 = LinearRegression()
lm2.fit(xdmgnd[['TEMP', 'TRNSPD']], xdmgnd['ACCDMG'])
print(f'Coefficients: {lm2.coef_}')
print(f'Intercept: {lm2.intercept_}')

# Three predictors
lm3 = LinearRegression()
lm3.fit(xdmgnd[['TEMP', 'TRNSPD', 'CARS']], xdmgnd['ACCDMG'])
print(f'Coefficients: {lm3.coef_}')
print(f'Intercept: {lm3.intercept_}')
```
