```python
# Model 1
xdmgnd_lm1 = ols('ACCDMG ~ TEMP + TRNSPD + CARS + HEADEND1', data=xdmgnd).fit()
print(xdmgnd_lm1.summary())

# Model 2
xdmgnd_lm2 = ols('ACCDMG ~ TEMP + TRNSPD + CARS', data=xdmgnd).fit()
print(xdmgnd_lm2.summary())
```

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf

def forward_selection(data, response, predictors):
    remaining = set(predictors)
    selected = []
    current_score, best_new_score = float('inf'), float('inf')
    formula_base = '{} ~ 1'.format(response)

    while remaining and current_score == best_new_score:
        scores_with_candidates = []
        for candidate in remaining:
            formula = '{} + {}'.format(formula_base, ' + '.join(selected + [candidate]))
            score = smf.ols(formula, data).fit().aic
            scores_with_candidates.append((score, candidate))
        scores_with_candidates.sort()
        best_new_score, best_candidate = scores_with_candidates[0]
        if current_score > best_new_score:
            remaining.remove(best_candidate)
            selected.append(best_candidate)
            current_score = best_new_score

    formula = '{} ~ {}'.format(response, ' + '.join(selected))
    model = smf.ols(formula, data).fit()
    return model

# Define predictors and response variable
predictors = ['TEMP', 'TRNSPD', 'CARS', 'HEADEND1']
response = 'ACCDMG'

# Apply forward selection
xdmgnd_lm1_fwd = forward_selection(xdmgnd, response, predictors)

# Display the summary of the final model
print(xdmgnd_lm1_fwd.summary()
```

```python
# Backward selection
def backward_selection(data, response, predictors):
    selected = predictors.copy()
    current_score, best_new_score = float('inf'), float('inf')
    formula_base = '{} ~ {}'.format(response, ' + '.join(selected))

    while selected and current_score == best_new_score:
        scores_with_candidates = []
        for candidate in selected:
            formula = '{} - {}'.format(formula_base, candidate)
            formula = formula.replace(' - +', ' +').replace(' - -', ' -')
            score = smf.ols(formula, data).fit().aic
            scores_with_candidates.append((score, candidate))
        scores_with_candidates.sort()
        best_new_score, worst_candidate = scores_with_candidates[0]
        if current_score > best_new_score:
            selected.remove(worst_candidate)
            current_score = best_new_score

    formula = '{} ~ {}'.format(response, ' + '.join(selected))
    model = smf.ols(formula, data).fit()
    return model

# Apply backward selection
xdmgnd_lm1_bwd = backward_selection(xdmgnd, response, predictors)

# Display the summary of the final model
print(xdmgnd_lm1_bwd.summary())
```

```python
# Partial F Test
anova_lm1_lm2 = sm.stats.anova_lm(xdmgnd_lm1, xdmgnd_lm2)
anova_lm1_fwd = sm.stats.anova_lm(xdmgnd_lm1, xdmgnd_lm1_fwd)
anova_lm1_bwd = sm.stats.anova_lm(xdmgnd_lm1, xdmgnd_lm1_bwd)

print(anova_lm1_lm2)
print(anova_lm1_fwd)
print(anova_lm1_bwd)
```

```python
# Test Sets
# Generate training sets and test sets from original data
test_size = 1/3
xdmgnd_train, xdmgnd_test = train_test_split(xdmgnd, test_size=test_size, random_state=42)

# Build models with training set
xdmgnd_lm1_train = smf.ols('ACCDMG ~ TEMP + TRNSPD + CARS + HEADEND1', data=xdmgnd_train).fit()
xdmgnd_lm2_train = smf.ols('ACCDMG ~ TEMP + TRNSPD + CARS', data=xdmgnd_train).fit()

# Predict with lm models
xdmgnd_lm1_pred = xdmgnd_lm1_train.predict(xdmgnd_test)
xdmgnd_lm2_pred = xdmgnd_lm2_train.predict(xdmgnd_test)

# Compute PMSE
pmse_xdmgnd_lm1 = mean_squared_error(xdmgnd_test['ACCDMG'], xdmgnd_lm1_pred)
pmse_xdmgnd_lm2 = mean_squared_error(xdmgnd_test['ACCDMG'], xdmgnd_lm2_pred)

print(pmse_xdmgnd_lm1, pmse_xdmgnd_lm2)
```

```python
# Version 2 Test Sets - Run multiple iterations with different testing and training sets
pmse1_result = []
pmse2_result = []

for _ in range(20):
    xdmgnd_train, xdmgnd_test = train_test_split(xdmgnd, test_size=test_size, random_state=np.random.randint(0, 1000))
    lm1_train = smf.ols('ACCDMG ~ TEMP + TRNSPD + CARS + HEADEND1', data=xdmgnd_train).fit()
    lm2_train = smf.ols('ACCDMG ~ TEMP + TRNSPD + CARS', data=xdmgnd_train).fit()

    lm1_pred = lm1_train.predict(xdmgnd_test)
    lm2_pred = lm2_train.predict(xdmgnd_test)

    pmse1_result.append(mean_squared_error(xdmgnd_test['ACCDMG'], lm1_pred))
    pmse2_result.append(mean_squared_error(xdmgnd_test['ACCDMG'], lm2_pred))

# Plot results
plt.plot(pmse1_result, label='Model 1 PMSE', color='blue')
plt.plot(pmse2_result, label='Model 2 PMSE', color='red')
plt.xlabel('Index')
plt.ylabel('PMSE')
plt.title('Model Comparison Based on PMSE')
plt.legend()
plt.show()
```

```python
# Paired t-test
from scipy.stats import ttest_rel, wilcoxon

t_stat, p_value = ttest_rel(pmse1_result, pmse2_result)
print('Paired t-test:', t_stat, p_value)

# Wilcoxon test
w_stat, w_p_value = wilcoxon(pmse1_result, pmse2_result)
print('Wilcoxon test:', w_stat, w_p_value)
```

```python
from scipy.stats import boxcox, boxcox_normmax, boxcox_normplot, boxcox_llf
from scipy.special import inv_boxcox

xdmgnd['ACCDMG'] = xdmgnd['ACCDMG'].replace(0, np.nan)
xdmgnd = xdmgnd.dropna(subset=['ACCDMG'])
xdmgnd['ACCDMG'] += 1e-6  # Add a small constant to avoid log(0)

# Generate a range of lambda values and apply Box-Cox transformation
lmbda_values = np.linspace(-2, 2, 100)
log_likelihoods = []

for lmbda in lmbda_values:
    transformed_data = boxcox(xdmgnd['ACCDMG'], lmbda=lmbda)
    log_likelihood = boxcox_llf(lmbda, xdmgnd['ACCDMG'])
    log_likelihoods.append(log_likelihood)

# Plot the log likelihoods for each lambda value
plt.plot(lmbda_values, log_likelihoods)
plt.xlabel('Lambda')
plt.ylabel('Log Likelihood')
plt.title('Log Likelihoods for Different Lambda Values')
plt.show()

# Choose the best lambda based on the plot
best_lambda = lmbda_values[np.argmax(log_likelihoods)]
print('Best lambda:', best_lambda)

# Apply Box-Cox transformation with the best lambda
xdmgnd['ACCDMG_boxcox'] = boxcox(xdmgnd['ACCDMG'], lmbda=best_lambda)

# Fit the model with Box-Cox transformed data
xdmgnd_lm1_boxcox = smf.ols(f'ACCDMG_boxcox ~ TEMP + TRNSPD + CARS + HEADEND1', data=xdmgnd).fit()

# Density plot after Box-Cox transformation
sns.kdeplot(xdmgnd['ACCDMG_boxcox'])
plt.title('Density plot after Box-Cox transformation')
plt.show()

# Diagnostic plots for Box-Cox model
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# Residuals vs Fitted
sns.residplot(x=xdmgnd_lm1_boxcox.fittedvalues, y=xdmgnd_lm1_boxcox.resid, lowess=True, ax=axs[0])
axs[0].set_title('Residuals vs Fitted')

# QQ plot
qqplot(xdmgnd_lm1_boxcox.resid, line='s', ax=axs[1])
axs[1].set_title('QQ plot')

# Scale-Location
sns.scatterplot(x=xdmgnd_lm1_boxcox.fittedvalues, y=np.sqrt(np.abs(xdmgnd_lm1_boxcox.resid)), ax=axs[2])
axs[2].set_title('Scale-Location')

plt.tight_layout()
plt.show()
```
