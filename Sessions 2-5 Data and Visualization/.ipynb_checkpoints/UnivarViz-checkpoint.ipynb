```python
# 1.4 Loading all years data

# You will need to read in all 20 years of the data
# You will put the data into a data structure called a list

# To do this you will use code I have written, AccidentInput.R
# Put that file in your working directory and then source it:
os.chdir(sourcedir)
# source("AccidentInput.R")  # R code, not required in Python

# Now use it to read in all the data. You must have ALL and ONLY the rail accident data
# files in one directory. Call that directory and its path, path.
# You can give your data a name
# In my examples below I use acts as the name for data sets
# Then use the command

os.chdir(traindir)
my_files = os.listdir(traindir)

# Get only csv files
my_csv_files = [file for file in my_files if file.endswith('.csv')]

# 2 different ways of loading multiple files in a directory

# The first uses the lapply function (equivalent in Python is using list comprehension)
acts = [pd.read_csv(file) for file in my_csv_files]

# The second uses file.inputl function from AccidentInput.R (not required in Python)
# acts = file.inputl(traindir)

# Now acts[0] is the data set from year 2001,
# acts[1] is the data set from year 2002, etc.
print(acts[0]['YEAR'])  # it shows 1 --> 2001
print(acts[4]['YEAR'])  # it shows 5 --> 2005

# Before we put all the data into one data frame
# we must clean the data
```

```python
# 2. Data Cleaning

# 2.1 columns in all years

# Are the number of columns different from year to year?
print(len(acts[0].columns))
print(len(acts[7].columns))
print(len(acts[6].columns))

# Get a common set the variables
common_vars = set(acts[0].columns).intersection(set(acts[7].columns))
print(common_vars)

# 2.1 combine all years of data after selecting common columns

# Now combine the data frames for all 20 years
# Use combine.data() functions which takes as arguments the files and the list
# of common variables
totacts = pd.concat([act[list(common_vars)] for act in acts], ignore_index=True)

# Check the number of rows and columns in this merged dataset
print(totacts.shape)

# The total accident should be a combination of all data
rows = sum([act.shape[0] for act in acts])
print(rows)  # total number of observations
print(totacts.shape[0])  # total number of observations

# How many entries (observations) does each year of accident have?
print(totacts.groupby('YEAR').size().reset_index(name='entries').sort_values(by='entries', ascending=False))
```

```python
# 3. Visualization

# 3.1 Histogram

# These examples are for year 2011
# Histograms are both in default way and also with ggplot2 package

# for 2011
data_2011 = acts[10]
plt.figure(figsize=(10, 6))
sns.histplot(data_2011['ACCDMG'], bins=30, kde=False, color='steelblue')
plt.title('Total Accident Damage in 2011')
plt.xlabel('Dollars ($)')
plt.ylabel('Frequency')
plt.show()

# Different bin widths

# Bin FD
plt.figure(figsize=(10, 6))
sns.histplot(data_2011['ACCDMG'], bins='fd', kde=False, color='steelblue')
plt.title('Total Accident Damage in 2011')
plt.xlabel('Dollars ($)')
plt.ylabel('Frequency')
plt.show()

# Bin Scott
plt.figure(figsize=(10, 6))
sns.histplot(data_2011['ACCDMG'], bins='scott', kde=False, color='steelblue')
plt.title('Total Accident Damage in 2011')
plt.xlabel('Dollars ($)')
plt.ylabel('Frequency')
plt.show()

# Bin 20
plt.figure(figsize=(10, 6))
sns.histplot(data_2011['ACCDMG'], bins=20, kde=False, color='steelblue')
plt.title('Total Accident Damage in 2011')
plt.xlabel('Dollars ($)')
plt.ylabel('Frequency')
plt.show()

# Bin 2
plt.figure(figsize=(10, 6))
sns.histplot(data_2011['ACCDMG'], bins=2, kde=False, color='steelblue')
plt.title('Total Accident Damage in 2011')
plt.xlabel('Dollars ($)')
plt.ylabel('Frequency')
plt.show()

# Other years
selected_years = totacts[totacts['YEAR'].isin([1, 4, 8, 11])]
plt.figure(figsize=(10, 6))
sns.histplot(data=selected_years, x='ACCDMG', bins=30, kde=False, color='steelblue')
plt.title('Total Accident Damage')
plt.xlabel('Dollars ($)')
plt.ylabel('Frequency')
plt.show()

# Accident damage 2009-2020

# Filter the selected years
selected_years = totacts[totacts['YEAR'].between(9, 20)]
# Histogram of accident damage based on each year
plt.figure(figsize=(10, 6))
sns.histplot(data=selected_years, x='ACCDMG', bins=30, kde=False, color='steelblue')
plt.title('Total Accident Damage in 2009-2020')
plt.xlabel('Dollars ($)')
plt.ylabel('Frequency')
plt.show()

# Histogram of accident damage based on each year with bins = 10
plt.figure(figsize=(10, 6))
sns.histplot(data=selected_years, x='ACCDMG', bins=10, kde=False, color='steelblue')
plt.title('Total Accident Damage in 2009-2020')
plt.xlabel('Dollars ($)')
plt.ylabel('Frequency')
plt.show()

# Damage in constant scales
selected_years = totacts[totacts['YEAR'].between(1, 11)]
# Histogram of accident damage based on each year
plt.figure(figsize=(10, 6))
sns.histplot(data=selected_years, x='ACCDMG', bins=30, kde=False, color='steelblue')
plt.title('Total Accident Damage in 2009-2020')
plt.xlabel('Dollars ($)')
plt.ylabel('Frequency')
plt.xlim(0, 1.7e7)
plt.ylim(0, 1000)
plt.show()
```

```python
# 3.3 Boxplots of ACCDMG
plt.figure(figsize=(10, 6))
sns.boxplot(x=data_2011['ACCDMG'], color='steelblue')
plt.title('Total Accident Damage in 2011')
plt.xlabel('Dollars ($)')
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(x=data_2011['ACCDMG'], color='steelblue')
plt.title('Total Accident Damage in 2011')
plt.xlabel('Dollars ($)')
plt.show()

# 4 plots on a single graph
selected_years = totacts[totacts['YEAR'].isin([1, 4, 8, 11])]
plt.figure(figsize=(10, 6))
sns.boxplot(x='YEAR', y='ACCDMG', data=selected_years, color='steelblue')
plt.title('Total Accident Damage in the selected years')
plt.xlabel('Dollars ($)')
plt.show()

# Box plot of accident damage based on each year
plt.figure(figsize=(10, 6))
sns.boxplot(x='YEAR', y='ACCDMG', data=selected_years, color='steelblue')
plt.title('Total Accident Damage in the selected years')
plt.xlabel('Dollars ($)')
plt.show()

# Box plot of accident damage based on each year with a constant scale
plt.figure(figsize=(10, 6))
sns.boxplot(x='YEAR', y='ACCDMG', data=selected_years, color='steelblue')
plt.title('Total Accident Damage in the selected years')
plt.xlabel('Dollars ($)')
plt.xlim(0, 1.7e7)
plt.show()

# Box plot of accident damage based on each year
selected_years = totacts[totacts['YEAR'].between(9, 19)]
# With a constant scale
plt.figure(figsize=(10, 6))
sns.boxplot(x='YEAR', y='ACCDMG', data=selected_years, color='steelblue')
plt.title('Total Accident Damage in the selected years')
plt.xlabel('Dollars ($)')
plt.show()

# Change the color for the outlier
plt.figure(figsize=(10, 6))
sns.boxplot(x=data_2011['ACCDMG'], color='steelblue')
plt.title('Total Accident Damage in 2011')
plt.xlabel('Dollars ($)')
plt.show()
```

```python
# 3.4 QQ plot of ACCDMG

# Does our data come from a Gaussian distribution
import scipy.stats as stats
plt.figure(figsize=(10, 6))
stats.probplot(data_2011['ACCDMG'], dist="norm", plot=plt)
plt.title('Total Accident Damage')
plt.show()

plt.figure(figsize=(10, 6))
stats.probplot(data_2011['TEMP'], dist="norm", plot=plt)
plt.title('Accident Temperature')
plt.show()
```

```python
# 3.4 Density plot of ACCDMG

d = data_2011['ACCDMG']
sns.kdeplot(d, shade=True, color='red')
plt.title('Gaussian kernel')
plt.show()

# Density plot on top of histogram
h = plt.hist(d, bins='fd', alpha=0.5, color='green', density=True)
sns.kdeplot(d, shade=True, color='red')
plt.title('Density and Histogram for ACCDMG')
plt.xlabel('ACCDMG')
plt.ylabel('Density')
plt.show()
```

```python
# 3.5 Bar plot of Accident Types

# Each number represents a different type of accident as given in the data dictionary,
# for e.g. 1-Derailment, 2-Head on Collision
plt.figure(figsize=(10, 6))
sns.countplot(x=data_2011['TYPE'], color='steelblue')
plt.title('Bar plot of Accident Type')
plt.xlabel('Accident type')
plt.show()
```
