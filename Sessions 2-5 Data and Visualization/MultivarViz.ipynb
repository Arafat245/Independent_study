```python
#******************************************************
#
#					    
#			 		Multivariate Viz
#
#******************************************************

import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#***************************
# 0.1 installing and loading the library for the visualization
#***************************
import sys
print(sys.version)
print(pd.__version__)
print(sns.__version__)

import warnings
warnings.filterwarnings("ignore")

#***************************
# 0.2 installing and loading the library for this session
#***************************
# The necessary libraries are already imported above.
```

```python
#**********************************************************
# 1. Reading in data
#**********************************************************

#***************************
# 1.1 Set working directory
#***************************

# Set working directory

# For example, a Mac user
traindir = "~/Google Drive/UVA/Courses/LSM/Fall2021/Data/TrainData/"
sourcedir ="~/Google Drive/UVA/Courses/LSM/Fall2021/Source/"

# or a Windows user
traindir = "C:/Users/jgh6ds/Desktop/Independent/Sessions 2-5 Data and Visualization/TrainData"
sourcedir = "C:/Users/jgh6ds/Desktop/Independent/Sessions 2-5 Data and Visualization"

# set the working directory to traindir
os.chdir(traindir)
# check the current working directory
# it should be the same as your traindir directory
print(os.getcwd())


#***************************
# 1.2 loading all years data
#***************************
# Source AccidentInput
os.chdir(sourcedir)
# AccidentInput.py contains similar functions to read and process data
from AccidentInput import file_inputl, combine_data

# You should have two data structures in working memory
# First - a list of data frames for each year of accident data
acts = file_inputl(traindir)

# Next, a data frame with all accidents from all years from 2001 - 2020
# with columns that are consistent for all of these years

# Get a common set the variables
comvar = set(acts[0].columns).intersection(set(acts[7].columns))

# the combined data frame
totacts = combine_data(acts, list(comvar))

# Get dimensions of the combined dataframe
print(totacts.shape)
```

```python
#*************************************************
#
#		2. More Data Cleaning
#
#*************************************************

# Variable names
print(totacts.columns)

# View the data types
print(totacts.dtypes)

# Look at the type for TYPE using describe
print(totacts['TYPE'].describe())

# Now, let's put more meaningful labels on TYPE variable
totacts['TYPE'] = totacts['TYPE'].astype('category')
totacts['TYPE'] = totacts['TYPE'].cat.rename_categories(["Derailment", "HeadOn", "Rearend", "Side", "Raking", "BrokenTrain", "Hwy-Rail", "GradeX", "Obstruction", "Explosive", "Fire","Other","SeeNarrative" ])

# Use value_counts() to see the frequencies
print(totacts['TYPE'].value_counts())
```

```python
#***********************************
#
# 	3. Visualization
#
#***********************************

#***************************
# 3.1 bar plot
#***************************
# Use barplot to graph this 

sns.set(style="whitegrid")

# Bar plot for Accident Frequency by Type
plt.figure(figsize=(10,6))
sns.barplot(x=totacts['TYPE'].value_counts().index, y=totacts['TYPE'].value_counts().values, palette="Blues_d")
plt.title('Accident Frequency by Type')
plt.xlabel('Type of Accident')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()

# Convert TYPEQ to numeric, errors='coerce' will convert non-numeric values to NaN
totacts['TYPEQ'] = pd.to_numeric(totacts['TYPEQ'], errors='coerce')

# Drop rows with NaN values in TYPEQ
totacts = totacts.dropna(subset=['TYPEQ'])

# Convert TYPEQ to category and rename categories
totacts['TYPEQ'] = totacts['TYPEQ'].astype('category')
totacts['TYPEQ'] = totacts['TYPEQ'].cat.rename_categories(["Freight", "Passenger", "Commuter", "Work", "Single", "CutofCars", "Yard", "Light", "Maint"])

# Bar plot for frequencies corresponding to different types of trains
plt.figure(figsize=(10,6))
sns.barplot(x=totacts['TYPEQ'].value_counts().index, y=totacts['TYPEQ'].value_counts().values, palette="Blues_d")
plt.xlabel('Type of Consist')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()

# Describe CAUSE
print(totacts['CAUSE'].describe())

# Create new variable 'Cause' from the first letter of 'CAUSE'
totacts['Cause'] = totacts['CAUSE'].apply(lambda x: x[0])

# Convert Cause to category
totacts['Cause'] = totacts['Cause'].astype('category')

# Value counts and bar plot for Cause
print(totacts['Cause'].value_counts())
plt.figure(figsize=(10,6))
sns.barplot(x=totacts['Cause'].value_counts().index, y=totacts['Cause'].value_counts().values, palette="Blues_d")
plt.xlabel('Cause')
plt.ylabel('Frequency')
plt.show()

# Histograms of TEMP with different breaks
plt.figure(figsize=(10,6))
sns.histplot(totacts['TEMP'], bins=range(-50, 121, 10))
plt.xlabel('Temperature')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(10,6))
sns.histplot(totacts['TEMP'], bins=range(-50, 121, 15))
plt.xlabel('Temperature')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(10,6))
sns.histplot(totacts['TEMP'], bins=range(-50, 121, 30))
plt.xlabel('Temperature')
plt.ylabel('Frequency')
plt.show()

# Histogram with changed color and title
plt.figure(figsize=(10,6))
sns.histplot(totacts['TEMP'], bins=range(-50, 121, 15), color='steelblue')
plt.title('Temperature Frequency')
plt.xlabel('Temperature')
plt.ylabel('Frequency')
plt.show()

# Drop columns from index 121 to 136
new_df = totacts.drop(totacts.columns[121:136], axis=1)
```

```python
# What's the narrative of the worst accident for total injured?

#***************************
# 3.1 scatter plot
#***************************

# without panel functions for 2020
sns.pairplot(acts[19][['TRKDMG', 'EQPDMG', 'ACCDMG', 'TOTINJ', 'TOTKLD']])
plt.show()

# with panel function- a little more detail
sns.pairplot(totacts[['TRKDMG', 'EQPDMG', 'ACCDMG', 'TOTINJ', 'TOTKLD']])
plt.show()

# Save as png to avoid problems in the document- make sure not to save in directory with data files
sns.pairplot(totacts[['TRKDMG', 'EQPDMG', 'ACCDMG', 'TOTINJ', 'TOTKLD']])
plt.savefig("metrics.png")
plt.close()

#***************************
# 3.2 Trellis Categorical Plots
#***************************
# Plotting damage per year
plt.figure(figsize=(10,6))
sns.boxplot(data=totacts, x='YEAR', y='ACCDMG')
plt.title('Box Plots of Accident Damage')
plt.xlabel('Year')
plt.ylabel('Damage ($)')
plt.show()

# Which accident has the most damage?
print(totacts[totacts['ACCDMG'] == totacts['ACCDMG'].max()])

# What type of accident had the most damage?
worst_dmg = totacts[totacts['ACCDMG'] == totacts['ACCDMG'].max()]
print(worst_dmg['TYPE'])

# Find out the worst accident for total killed and injured
max_totinj = totacts['TOTINJ'].idxmax()
print(totacts.loc[max_totinj, ['TOTINJ', 'TYPE', 'ACCDMG']])

# What's the narrative of the worst accident for total injured?
worst_inj = totacts[totacts['TOTINJ'] == totacts['TOTINJ'].max()]
print(worst_inj.loc[:, worst_inj.columns.str.startswith('NARR')])
print(worst_inj['ACCDMG'])

# Find the worst accidents in 2018. What happened?
acts_2018 = totacts[totacts['YEAR'] == 18]

# damage
print(acts_2018.loc[acts_2018['ACCDMG'].idxmax()])
# injury
print(acts_2018.loc[acts_2018['TOTINJ'].idxmax()])
# killed
print(acts_2018.loc[acts_2018['TOTKLD'].idxmax()])

# Find the accidents with the most injuries in 2018. What happened in these accidents?
# Explore the news to understand underlying causes and what could have been done to prevent these accidents.
print(acts_2018.nlargest(5, 'TOTINJ'))

# Find the accident with the most damage in 2018. What do you notice about the accidents with the most damage and those with the most injuries?
print(acts_2018.nlargest(5, 'ACCDMG'))

# Plotting accident cause vs. damage
plt.figure(figsize=(10,6))
sns.boxplot(data=totacts, x='Cause', y='ACCDMG')
plt.title('Box Plots of Accident Damage')
plt.xlabel('Accident Cause')
plt.ylabel('Damage ($)')
plt.show()

# Plot scaled (log) accident damage
plt.figure(figsize=(10,6))
sns.boxplot(data=totacts, x='Cause', y='ACCDMG', showfliers=False)
plt.yscale('log')
plt.title('Box Plots of Log(Accident Damage)')
plt.xlabel('Accident Cause')
plt.ylabel('log(Damage ($))')
plt.show()

# Plot cause vs. no. killed 
plt.figure(figsize=(10,6))
sns.boxplot(data=totacts, x='Cause', y='TOTKLD')
plt.title('Box Plots of Total Killed')
plt.xlabel('Accident Cause')
plt.ylabel('Total Killed')
plt.show()

# Plot cause vs. injured
plt.figure(figsize=(10,6))
sns.boxplot(data=totacts, x='Cause', y='TOTINJ')
plt.title('Box Plots of Total Injured')
plt.xlabel('Accident Cause')
plt.ylabel('Total Injured')
plt.show()

# X-Y plots conditioned on cause for both killed and injured
g = sns.FacetGrid(totacts, col='Cause', col_wrap=4, sharex=False, sharey=False)
g.map(sns.scatterplot, 'ACCDMG', 'TOTKLD')
g.add_legend()
g.set_axis_labels('Total Accident Damage', 'Total Killed')
g.fig.subplots_adjust(top=0.9)
g.fig.suptitle('Damage vs. Killed Conditioned on Cause')
plt.show()
```
